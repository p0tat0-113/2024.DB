```python
import pandas as pd
import sqlite3

data = pd.read_csv('./menu.csv')
conn = sqlite3.connect('McDonalds.db')
data.to_sql('MCDONALDS_NUTRITION', conn)#또 실행시키니까 ValueError: Table 'MCDONALDS_NUTRITION' already exists.예외가 발생함.
```

```python
#pandas를 통해 sql명령어를 보낸다. df: data frame
df = pd.read_sql("select * from MCDONALDS_NUTRITION ", conn)
df.head()#5줄을 출력하게 해줌.
```

```python
df.describe(include='all')#테이블에 대한 명세서를 찍어본다.
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#범주형 스캐터 플롯
plot = sns.swarmplot(x = "Category", y = "Sodium", data = df)#x축은 범주, y축은 나트륨 함량, data는 아까 가져온 df
plt.setp(plot.getxticklabels(), rotation = 70)
plt.title('Sodium Content')
plt.show()
```
